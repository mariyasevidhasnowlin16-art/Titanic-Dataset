```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
data.head()
```

```python

age_series = data["Age"]
fare_series = data["Fare"]
survival_series = data["Survived"]

print(type(age_series))
print(type(data))
```

```python
data.head()
data.info()
data.describe()
```

```python
data.isnull().sum()

data["Age"].fillna(data["Age"].mean(), inplace=True)

data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)
data.dropna(inplace=True)
```

```python
data.loc[0]
data.loc[0, "Name"]
```

```python
data.iloc[0:5]
data.iloc[:, 2:5]
```

```python
data.rename(columns={"Sex": "Gender"}, inplace=True)
```

```python
pivot_table = data .pivot_table(values="Survived",
                             index="Gender",
                             columns="Pclass",
                             aggfunc="mean")
print(pivot_table)
```

```python
data[data["Age"] > 30]

data[data["Fare"] > 50]

data[(data["Gender"] == "female") & (data["Survived"] == 1)]
data.sort_values(by="Fare", ascending=False)
data.sort_values(by="Age")
data.sort_values(by="Survived")
```

```python
data.groupby("Gender")["Survived"].mean()
data.groupby("Pclass")["Age"].mean()
data.groupby("Embarked")["Survived"].sum()
```

```python
class_info = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Name": ["First", "Second", "Third"]
})
merged_data = pd.merge(data, class_info, on="Pclass")
combined = pd.concat([data.head(), data.tail()])
```
